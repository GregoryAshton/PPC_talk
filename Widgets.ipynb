## Lorenz attractor


```python
def lorenz(ys, t, s, r, b) :
    " Defined the ODEs for the Lorenz attractor "
    [x, y, z] = ys
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return x_dot, y_dot, z_dot

time = np.linspace(0, 100, 10000)
results = odeint(lorenz, y0=[0.0, 1.0, 1.05], t=time, args=(10, 28, 2.667))

plt.plot(results[:, 0], results[:, 2])
plt.show()
```

## Interactive Lorenz attractor

Now we will make it 3D and interative

```python
from mpl_toolkits.mplot3d import Axes3D

def solve_lorenz(s, r, b):
    def lorenz(ys, t, s, r, b) :
        [x, y, z] = ys
        x_dot = s*(y - x)
        y_dot = r*x - y - x*z
        z_dot = x*y - b*z
        return x_dot, y_dot, z_dot

    time = np.linspace(0, 100, 10000)
    results = odeint(lorenz, y0=[0.0, 1.0, 1.05], t=time, args=(s, r, b))

    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.plot(results[:, 0], results[:, 1], results[:, 2])
    plt.show()

```

```python
from IPython.html.widgets import interact, interactive, fixed

interactive(solve_lorenz, s=10, r=28, b=2.667)
```
